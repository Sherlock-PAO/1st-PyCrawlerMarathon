### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)
soup = BeautifulSoup(r.text,'html5lib')

news_today = soup.find(class_='part_list_2').find_all('h3')
#print(news_today[0])
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
news_dict={}
for news in news_today:
    #print(news.find('em')['class'])
    if news.find('em')['class'][-1] not in news_dict:
        news_dict[news.find('em')['class'][-1]]=[]
    news_dict[news.find('em')['class'][-1]].append(news.find('a').text)
#print(news_dict)
for key in news_dict.keys():
    print(key)
    print('**************************************************')
    for n in news_dict[key]:
        print(n)
    print()
```

### ③ 哪一個類別的文章最多

```python
def maxInDict(dict_):
    maxlen = 0
    for key in dict_.keys():
        #print(key,'的文章有',len(dict_[key]))
        if len(dict_[key])>maxlen:
            maxlen = len(dict_[key])
            flag = key
    print(flag,'的文章最多')
maxInDict(news_dict)
```
