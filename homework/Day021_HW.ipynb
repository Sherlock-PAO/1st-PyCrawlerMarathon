## 作業目標

根據範例：

1. 取出今天所有的新聞
2. 取出現在時間兩小時內的新聞

## 1. 取出今天所有的新聞

```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time
from datetime import datetime,timedelta
import pandas as pd
```

```python
yesterday_date = datetime.now()-timedelta(days=1)
yesterday_str = yesterday_date.strftime("%Y/%m/%d")
print(yesterday_str)
```

```python
browser = webdriver.Chrome(executable_path='chromedriver')
browser.maximize_window()
browser.get('https://www.ettoday.net/news/news-list.htm')
while True:
    browser.execute_script("window.scrollTo(0,document.body.scrollHeight);")
    time.sleep(2)
    html_source = browser.page_source
    soup = BeautifulSoup(html_source,'html5lib')
    last_news_time = soup.find_all(class_="part_list_2")[-1].find_all('h3')[-1].find(class_="date").text
    print(last_news_time)
    if yesterday_str in last_news_time:
        print('====STOP====')
        break
browser.close()
```

```python
today_news = []
for d in soup.find(class_='part_list_2').find_all('h3'):
    if yesterday_str in d.find(class_="date").text:
        break
    today_news.append({
        '時間': d.find(class_="date").text,
        '內容': d.find_all('a')[-1].text
    })
pd.DataFrame(today_news)
```

## 2. 取出現在時間兩小時內的新聞

```python
past_2hours_news=[]
two_hours_ago = datetime.now() - timedelta(hours=2)

for d in soup.find(class_="part_list_2").find_all('h3'):
    if datetime.strptime(d.find(class_="date").text, '%Y/%m/%d %H:%M') < two_hours_ago:
        break
    past_2hours_news.append({
        '時間': d.find(class_="date").text,
        '內容': d.find_all('a')[-1].text
    })
pd.DataFrame(past_2hours_news)
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
three_days_ago = datetime.now() - timedelta(days=3)
print('三天前:',three_days_ago)

three_days_ago_month = str(int(three_days_ago.strftime('%m')))

print('三天前(月):',three_days_ago_month)

three_days_ago_day = str(int(three_days_ago.strftime('%d')))
print('三天前(日):',three_days_ago_day)

three_days_15_clock = datetime.strptime(three_days_ago.strftime('%Y/%m/%d')+' 15','%Y/%m/%d %H')
print('三天前下午三點',three_days_15_clock)
three_days_17_clock = datetime.strptime(three_days_ago.strftime('%Y/%m/%d')+' 17','%Y/%m/%d %H')
print('三天前下午五點',three_days_17_clock)
```

```python
from selenium.webdriver.support.ui import Select
browser = webdriver.Chrome(executable_path='chromedriver')
browser.maximize_window()
browser.get('https://www.ettoday.net/news/news-list.htm')

selM = Select(browser.find_element_by_id("selM"))
selM.select_by_value(three_days_ago_month)
selD = Select(browser.find_element_by_id("selD"))
selD.select_by_value(three_days_ago_day)
browser.find_element_by_id('button').click()
```

```python
while True:
    browser.execute_script("window.scrollTo(0,document.body.scrollHeight);")
    time.sleep(2)
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, "html5lib")
    last_news_time = soup.find_all(class_="part_list_2")[-1].find_all('h3')[-1].find(class_="date").text
    print(last_news_time)
    if datetime.strptime(last_news_time, '%Y/%m/%d %H:%M') < three_days_15_clock:
        print('====STOP====')
        break
browser.close()
```

```python
three_days_ago_news=[]
for d in soup.find(class_="part_list_2").find_all('h3'):
    if datetime.strptime(d.find(class_="date").text, '%Y/%m/%d %H:%M') > three_days_17_clock:
        continue
    elif datetime.strptime(d.find(class_="date").text, '%Y/%m/%d %H:%M') < three_days_15_clock:
        break
    three_days_ago_news.append({
        '時間': d.find(class_="date").text,
        '內容': d.find_all('a')[-1].text
    })
pd.DataFrame(three_days_ago_news)
```
